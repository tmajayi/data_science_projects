```python
import numpy as np
import scipy.io as sio
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA


data = sio.loadmat('ex7data1.mat')
X = data['X']
x1,x2 = X[:,0],X[:,1]

pca = PCA(n_components=1)
model = pca.fit(X)
Xr = model.transform(X)
xn = model.inverse_transform(Xr)  #projection
plt.plot(xn[:,0],xn[:,1],'b.',label='pca projection')
plt.plot(x1,x2,'r.',label='raw')
plt.legend()
plt.show()
```

```python
faces = sio.loadmat('ex7faces.mat')
X = faces['X']
from random import sample
n = 100
l = int(np.sqrt(n))
smpl = sample(range(len(X)),n)
Xs = X[smpl,:]

fig = plt.figure(figsize=(8, 6), dpi=120, facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace=0.04, wspace=-0.78)
for i in range (1,n+1):
    ax = fig.add_subplot(l,l,i)
    xs = Xs[i-1].reshape(32,32)
    ax.imshow(xs.T,cmap='gray')
    plt.axis('off')
    
plt.show()
```

```python
nc = 100    #reduce 1024 data to 100 principal components
pca2 = PCA(n_components=nc)
Xpc = pca2.fit_transform(X)
Xt = pca2.inverse_transform(Xpc)  #reconstruct faces from 100 principal components
Xs = Xt[smpl,:]

fig = plt.figure(figsize=(8, 6), dpi=120, facecolor='w', edgecolor='k')
fig.subplots_adjust(hspace=0.04, wspace=-0.78)
for i in range (1,n+1):
    ax = fig.add_subplot(l,l,i)
    xs = Xs[i-1].reshape(32,32)
    ax.imshow(xs.T,cmap='gray')
    plt.axis('off')
```
