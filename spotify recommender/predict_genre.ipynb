```python
features = df_gen.drop(['genres'], axis=1).values
features.shape
```

```python
from sklearn.preprocessing import StandardScaler

feat_norm = StandardScaler().fit_transform(features)

#def predict_genre(X):
df = pd.read_csv('data\data_w_genres.csv')
X = df.iloc[:5,2:15].values
X = StandardScaler().fit_transform(X)
predict = []
tol = 0.1
try:
    m,n = X.shape
    for i in range(m):    
        dist = np.linalg.norm(feat_norm - X[i], axis=1)
        genre = df_gen.genres.iloc[np.where(dist <= (min(dist)+tol))].tolist()
        predict.append(genre)
        
except ValueError:  # if X is a 1D array
    dist = np.linalg.norm(features - X, axis=1)
    predict = df_gen.genres.iloc[np.where(dist < 1)].tolist()

print(predict)
#return predict

#predict_genre(feat_norm[23])
```
